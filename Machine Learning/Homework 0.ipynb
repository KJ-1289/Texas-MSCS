**2. How many malignant cases are there in total?**

```python
target = pd.DataFrame(cancer.target)
target.rename(columns={0:'benign'}, inplace=True)
#A value of 0 is malignant
#A value of 1 is benign
target.value_counts()
```

**3. Make a scatterplot of ‘mean compactness’ and ‘mean perimeter’. Try to color the points by their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot (from seaborn) useful for this.**

```python
df2 = pd.concat([target, df], axis=1)
df2.head()
```

```python
sns.jointplot(data=df2, x='mean compactness', y='mean perimeter', hue='benign')
```

**4. Make pairwise scatterplots of ‘mean texture’, ‘mean perimeter’, ‘mean compactness’, and ‘mean radius’. You may find scatter matrix (from matplotlib) or pairplot (from seaborn) useful for this. Do you notice that two of these features seem to be particularly correlated? Why might this be?**

```python
df3 = df2[['benign', 'mean texture', 'mean perimeter', 'mean compactness', 'mean radius']]
sns.pairplot(df3, hue='benign')
```
